```python
# value in each group
labels = ['6_to_18', '65_and_older', '18_to_24', '45_to_54', '35_to_44', '25_to_34', '55_to_64', '5_and_Under']
values = [1615, 1111, 900, 910, 858, 1336, 738, 2312]

# Graph it
plt.figure(figsize=(10, 6))
plt.bar(labels, values)
plt.xlabel('Group')
plt.ylabel('Number of Pictures')
plt.title('Number of Pictures in Each Group')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# values
labels = ['5_and_Under', '6_to_18', '18_to_24', '25_to_34', '35_to_44', '45_to_54', '55_to_64', '65_and_older']
original_values = [2312, 1615, 900, 1336, 858, 910, 738, 1111]

# values
rotated_values = [100, 600, 900, 900, 858, 910, 738, 1000]

# stacked bar graph
plt.figure(figsize=(10, 6))
plt.bar(labels, original_values, label='original')
plt.bar(labels, rotated_values, bottom=original_values, label='rotated')
plt.xlabel('Group')
plt.ylabel('Number of Pictures')
plt.title('Number of Pictures in Each Group')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Original values
labels = ['5_and_Under', '6_to_18', '18_to_24', '25_to_34', '35_to_44', '45_to_54', '55_to_64', '65_and_older']
original_values = [2312, 1615, 900, 1336, 858, 910, 738, 1111]

# Rotated values
rotated_values = [100, 600, 900, 900, 858, 910, 738, 1000]

# Flipped values
flipped_values = [200, 397, 812, 376, 896, 792, 1136, 501]

# Create a stacked bar graph
plt.figure(figsize=(10, 6))
plt.bar(labels, original_values, label='Original')
plt.bar(labels, rotated_values, bottom=original_values, label='Rotated')
plt.bar(labels, flipped_values,
        bottom=[i+j for i,j in zip(original_values, rotated_values)], label='Flipped')
plt.xlabel('Group')
plt.ylabel('Number of Pictures')
plt.title('Number of Pictures in Each Group')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
